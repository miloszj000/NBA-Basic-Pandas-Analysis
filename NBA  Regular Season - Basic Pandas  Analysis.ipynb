# CONVERTING TYPES

```python
stats['Pos'] = stats['Pos'].astype('category')
stats['Tm'] = stats['Tm'].astype('category')
```

```python
stats['Tm'].unique()
```

```python
stats = stats.loc[stats['Tm'] != 'TOT']
```

```python
stats.info(memory_usage='deep')
```

# INDEX

```python
stats.set_index(['Tm','Player','Pos'], inplace = True)
```

```python
stats =  stats.sort_index(level = 'Tm')
```

# CLEANING DATA

```python
stats.dropna(inplace=True)
```

```python
stats.head()
```

```python
stats = stats.where(stats['G'] >= 10)
```

# BEST SHOOTERS OVR

```python
stats.dropna(inplace=True)
```

```python
pts_mean = stats['PTS'].mean()
pts_max = stats['PTS'].max()
fgproc_mean = stats['FG%'].mean()
fgproc_max = stats['FG%'].max()
```

# TOP 5 SHOOTERS BY PTS

```python
pts_top = stats.where((stats['PTS'] > (pts_max-pts_mean))).dropna().sort_values(by='PTS',ascending=False).head(15)
pts_top
```

```python
fg_top = stats.where((stats['FG%'] < 0.99))
fg_top.sort_values(by='FG%', ascending=False)
```

# TOP 5 SHOOTERS BY FG%

```python
fg_top = stats.where((stats['FG%'] < 0.99))
```

# TOP SHOOTERS BY FG% AND PTS

```python
top_shooters = pts_top.merge(fg_top, on='Player',how='inner',suffixes=[' ',' '])
top_shooters = top_shooters.iloc[0:8 , [2,7,26]]
top_shooters
```

# MVP RACE

## To calcuate the best players in the league, I'm adding up all vital stats and substracting negative stats.

```python
stats['MVP_SCORE'] = (stats['FG'] + stats['FT%'] + stats['TRB'] + stats['AST'] + stats['STL'] + stats['BLK'] - stats['TOV'] - stats['PF'])
```

```python
MVP = stats.sort_values(by='MVP_SCORE', ascending = False).iloc[0:10 , [1,2,27]]
```
